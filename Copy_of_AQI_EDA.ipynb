```python
import pandas as pd #for importing data
import plotly.express as px 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objs as go
```

```python
cities=pd.read_csv('https://raw.githubusercontent.com/RoshanKrSharma/AQI-Data-analysis/main/city_day.csv')
cities
```

```python
cities.head()
```

```python
missing= pd.DataFrame(cities.isna().sum()) #counting NaN in each column of cities dataframe and storing the count in a new dataframe
```

```python
missing.columns=['count']
missing
```

```python
cities.fillna(0,inplace=True) #handling null values
```

```python
cities.head()
```

```python
cities.info()
```

```python
cities['Date']=pd.to_datetime(cities['Date']) #changing 'Date' to date-time format
```

```python
cities.info()
```

```python
cities['month']=pd.DatetimeIndex(cities['Date']).month #making a new column by extracting month part of the Date column in cities dataframe
cities['Year']=pd.DatetimeIndex(cities['Date']).year ##making a new column by extracting year part of the Date column in cities dataframe
```

  Pandas DatetimeIndex.date attribute outputs an Index object containing the date values present in each of the entries of the DatetimeIndex object. 

Now using this index we can easily data of particular year or particular range. 

Syntax: DatetimeIndex.date

Return: numpy array of python datetime.date 

```python
cities.head()
```

Now, we'll merge columns of PM2.5 & PM10 and make a new column named PM

Similarly, merging NO,NO2,NOx and creating new column named Nitric

Similarly, merging Benzene, toluene,xylene and creating new column named BTX

```python
cities['PM']=cities['PM10']+cities['PM2.5']
cities['Nitric']=cities['NO']+cities['NO2']+cities['NOx']
cities['BTX']=cities['Benzene']+cities['Toluene']+cities['Xylene']
```

```python
cities.head()
```

```python
cities_group=cities.groupby(['City','month','Year'])[['PM','Nitric','BTX','NH3','CO','SO2','O3','AQI']].mean()
```

```python
cities_group
```

```python
cities_group=cities_group.reset_index(['City','month','Year'])
```

```python
cities_group
```

 **Now we will visualize and see in which city which parameter of pollution is more, etc**

```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

**plotting top 10 cities vs pollutants**

```python
cities_15_19=cities_group[cities_group['Year'].isin([2017,2016,2015,2018,2019])]

#list of pollutants
pollutants=['PM','Nitric','CO','NH3','O3','SO2','BTX','AQI']
sns.set_theme(style='whitegrid')

#plotting the top 10 most polluted cities for each category of pollutants, as well as overall AQI
for i in pollutants:
  df=cities_15_19.groupby(['City'])[[i]].mean().sort_values(i,ascending=False).iloc[:10,:]
  df=df.reset_index(['City'])
  plt.figure()
  sns.barplot(data=df,x='City',y=i,palette='viridis',alpha=.9)
  plt.xticks(rotation=45)
```

**Now we'll see that which pollutant(like PM,Nitric, etc) contributes more on AQI**

**for this we'll have to find correlation between pollutants and AQI**

**correlation coefficient range is -1 to 1**

```python
#correlation matrix
corr_matrix=cities_group.iloc[:,3:].corr()
print(corr_matrix)
```

**Now we'll see yearly trend of air quality in each cities**

```python
df_AQI_trend=cities_15_19.groupby(['City','month'])[['AQI']].mean().reset_index()
fig=px.line(df_AQI_trend,x="month",y="AQI",color="City",line_shape='spline')#line_shape=spline se line smooth hota h
fig.show()
```

**Now we will see difference of AQI in 2019 and march 2020(lockdown started from march 2020)**

*as we know full lockdown was there in India from march 2020, so there must be significant decrease in AQI*


```python
cities_19_all=cities_group[cities_group['Year']==2019]
cities_20_all=cities_group[cities_group['Year']==2020]
cities_19_vs_cities_20_all=pd.merge(cities_19_all,cities_20_all,how="inner",on=["City","month"])
months=['Jan','Feb','Mar','Apr','May','Jun','Jul']

fig=go.Figure()
fig.add_trace(go.Bar(
    x=months,
    y=cities_19_all[cities_19_all["City"]=="Delhi"]["AQI"],
    name='2019',
    marker_color='indianred'

))

fig.add_trace(go.Bar(
    x=months,
    y=cities_20_all[cities_20_all["City"]=="Delhi"]["AQI"],
    name='2020',
    marker_color='lightsalmon'
))

fig.update_layout(
    title='AQI comparision Delhi',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title='Air Quality Index(AQI)',
        titlefont_size=16,
        tickfont_size=14
    ),
)
```

**Now we'll analyse which pollutant was most decreased in year 2020 in comparision to 2019**

```python
cities_=['Ahmedabad','Lucknow','Visakhapatnam','Jaipur']
#understanding the rise of pollutants which contributed to the decreased AQI in 2020 by comparing the levels of each pollutant
for i in cities_:
  city_19_20=cities_group[(cities_group['City']==i)&(cities_group['Year'].isin([2019,2020])) & (cities_group['month']<8)]
  sns.set_theme(style='whitegrid')
  fig=plt.figure()
  fig,axes=plt.subplots(2,4,figsize=(12,5))

  sns.barplot(
      data=city_19_20,
      x='month',y='AQI',hue='Year',
      palette='dark', alpha=.6,ax=axes[0,0]
  )

  sns.lineplot(
      data=city_19_20,
      x='month',y='PM', hue='Year',palette='dark',
      markers=True,dashes=False,ax=axes[0,1]
  )
  
  sns.lineplot(
      data=city_19_20,
      x='month', y='Nitric', hue='Year',palette='dark',
      markers=True,dashes=False,ax=axes[0,2]
  )

  sns.lineplot(
      data=city_19_20,
      x='month',y='CO',hue='Year',palette='dark',
      markers=True,dashes=False,ax=axes[0,3]
  )
  sns.lineplot(
      data=city_19_20,
      x='month',y='BTX',hue='Year',palette='dark',
      markers=True,dashes=False,ax=axes[1,0]
  )
  sns.lineplot(
      data=city_19_20,
      x='month',y='SO2',hue='Year',palette='dark',
      markers=True,dashes=False,ax=axes[1,1]
  )
  sns.lineplot(
      data=city_19_20,
      x='month',y='NH3',hue='Year',palette='dark',
      markers=True,dashes=False,ax=axes[1,2]
  )
  sns.lineplot(
      data=city_19_20,
      x='month',y='O3',hue='Year',palette='dark',
      markers=True,dashes=False,ax=axes[1,3]
  )
  fig.tight_layout()
  print(i,':')
  plt.show()

  
```
